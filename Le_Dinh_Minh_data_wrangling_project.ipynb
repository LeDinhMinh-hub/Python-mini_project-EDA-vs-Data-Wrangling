```python
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv('prices.csv')
sales = pd.read_csv('sales.csv')
```

```python
prices.head()
```

```python
sales.head()
```

```python
from datetime import datetime
```

```python
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'], format="%m/%d/%y %H:%M")
```

```python
prices['updated_at'] = pd.to_datetime(prices['updated_at'], format="%m/%d/%y %H:%M")
```

```python
prices.head()
```

```python
sales.head()
```

```python
sales = sales.sort_values('product_id')
```

```python
merge_table = pd.merge_asof(sales, prices, on='product_id', direction='nearest')
```

```python
merge_table
```

```python
merge_table['price'] = merge_table.apply(lambda row: row['new_price'] if row['ordered_at'] >= row['updated_at'] else row['old_price'], axis=1)
```

```python
merge_table
```

```python
merge_table['revenue'] = merge_table['quantity_ordered'] * merge_table['price']
```

```python
merge_table
```

```python
grouped_revenue = merge_table.groupby('product_id').agg({'revenue': 'sum'})
```
